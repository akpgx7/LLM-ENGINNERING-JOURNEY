```python
from openai import OpenAI
from scraper import fetch_website_contents

ollama = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

system_prompt = """
    You are an helpful Assistant,
    you summarize websites, ignore navigation junk, respond in markdown
"""
def user_prompt_for(coursera):
    user_prompt = "Hello , I need your help to summarize this given website for me"
    user_prompt += coursera
    return user_prompt
def messages_for(website):
    return [{'role':'system','content': system_prompt},
    {'role':'user','content':user_prompt_for(website)}]


coursera_text = fetch_website_contents("https://www.coursera.org/")
def summarize(coursera):
    response = ollama.chat.completions.create(
    model="gemma3",
    messages=messages_for(coursera))
    return response.choices[0].message.content

print(summarize(coursera_text))
```
